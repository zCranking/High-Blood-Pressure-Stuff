```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("cardiovascular.csv")
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of itd
groupby_ageC = df.groupby('age')['cholesterol'].mean().reset_index()
groupby_ageC
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
fig = plt.subplots(figsize=(10,8))
value = groupby_ageC['age']
label = groupby_ageC['cholesterol']
plt.plot(label, value, label="cholesterol", linewidth=3.0)
plt.xlabel("Cholesterol")
plt.ylabel("Age")
plt.title("Line Graph of Age VS Cholesterol")
plt.legend()
plt.show()
```

# Correlation between systolic and diastolic blood pressure

```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats
df
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
groupby_ageS = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
groupby_ageS
```

```python
fig = plt.subplots(figsize=(12,8))
label = groupby_ageS['systolic_blood_pressure']
value = groupby_ageS['age']
plt.plot(label, value, label="Systolic Blood Pressure", linewidth=3.0)
plt.xlabel("Systolic Blood Pressure")
plt.ylabel("Age")
plt.title("Line Graph of Age VS Systolic Blood Pressure")
plt.legend()
plt.show()
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it
groupby_ageD = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
groupby_ageD
```

```python
fig = plt.subplots(figsize=(12,8))
label = groupby_ageD['diastolic_blood_pressure']
value = groupby_ageD['age']
plt.plot(label, value, label="Diastolic Blood Pressure", linewidth=3.0)
plt.xlabel("Diastolic Blood Pressure")
plt.ylabel("Age")
plt.title("Line Graph of Age VS Diastolic Blood Pressure")
plt.legend()
plt.show()
```
